```python
# fetch the earthquake catalog from the IRIS 
from obspy.clients.fdsn import Client
from obspy.core import UTCDateTime
client = Client("IRIS")

t1 = UTCDateTime("2008-01-01T00:00:00")
t2 = UTCDateTime("2018-01-08T23:59:59")

# set the latitude and longitude to the center of your array
catalog = client.get_events(starttime=t1, endtime=t2, minradius=20, latitude=44.5, longitude=-110.5, maxradius=100, minmagnitude=5.5, maxmagnitude = 8)
```

```python
# catalog.write("./catalog/catalog.xml", format="QUAKEML") 
print(catalog)
```
